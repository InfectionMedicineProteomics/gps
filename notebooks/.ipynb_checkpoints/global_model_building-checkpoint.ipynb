```python
class TooManyNodesException(Exception):

    pass


class Node:

    def __init__(self, key='', data=None, edges=[], ntype=None):

        self.data = data

        self.key = key

        self.edges = edges

        self.ntype = ntype


class Edge:

    def __init__(self, nodes=[], weight=1.0):

        if len(nodes) > 2:

            raise TooManyNodesException(
                'No more than 2 nodes can be connected by an edge'
            )

        self.nodes = nodes

        self.weight = weight


class Graph:

    def __init__(self):

        self.nodes = dict()

    def connect_nodes(self, node_a, node_b, weight=1.0, directed=False):

        if directed:

            pass

        else:

            edge = Edge(
                nodes=[
                    node_a,
                    node_b
                ]
            )

            print('here')
            self.nodes[node_a].edges.append(
                edge
            )

            self.nodes[node_b].edges.append(
                edge
            )
    
    def select_nodes_by_type(self, ntype=''):

        return_nodes = dict()

        for key, item in self.nodes.items():

            if item.ntype == ntype:

                return_nodes[key] = item

        return return_nodes
            
```

```python
class PeptideNode:

    def __init__(self, peptides):

        self.peptides = peptides

    def get_proteins(self):

        proteins = list(
            set(
                self.peptides['protein_accession']
            )
        )

        return proteins
```

```python
def iterate_bipartite(graph, side=''):

    for node, attributes in graph.nodes(data=True):

        if attributes['bipartite'] == side:

            yield node
```

```python
protein_roll_up_model = nx.Graph()

peptides = dict()

proteins = list()

for osw_path in osw_paths:

    peak_groups = fetch_peak_groups(
        host=osw_path,
        query=FETCH_SCORED_DATA
    )

    highest_ranking = peak_groups.select_peak_group(
        rank=1,
        rerank_keys=['alt_d_score'],
        ascending=False
    )

    for record_idx, record in highest_ranking.iterrows():

        peptide_sequence_charge = f"{record['peptide_sequence']}_{record['charge']}"

        record['peptide_sequence_charge'] = peptide_sequence_charge

        if peptide_sequence_charge not in peptides:

            peptides[peptide_sequence_charge] = list()

        peptides[peptide_sequence_charge].append(
            record
        )

        proteins.append(record['protein_accession'])
    
proteins = list(set(proteins))

peptide_nodes = list()

for peptide_id, records in peptides.items():

    peptide_nodes.append(
        {
            'peptide_id': peptide_id,
            'peptide_records': records
        }
    )
    
protein_roll_up_model.add_nodes_from(proteins, bipartite='proteins')

peptide_nodes = list()

for peptide_id, records in peptides.items():

    converted_records = PeptideNode(
        peptides=pd.DataFrame(records)
    )

    peptide_nodes.append(converted_records)
    
protein_roll_up_model.add_nodes_from(peptide_nodes, bipartite='peptides')

for peptide_node in iterate_bipartite(protein_roll_up_model, side='peptides'):
    for protein in peptide_node.get_proteins():

        protein_roll_up_model.add_edge(
            peptide_node,
            protein
        )
        
combined_scores = list()

for peptide_node in iterate_bipartite(protein_roll_up_model, side='peptides'):

    peptide_group = dict(
        peptide_node.peptides.sort_values(
            'alt_d_score', 
            ascending=False
        ).iloc[0].copy()
    )
    
    peptide_mean_score = peptide_node.peptides.alt_d_score.mean()

    peptide_mean_vote = peptide_group['vote_percentage']

    peptide_id = peptide_group['peptide_sequence_charge']

    protein_id = peptide_group['protein_accession']

    combined_scores.append(
        {
            'peptide_id': peptide_id,
            'protein_id': protein_id,
            'alt_d_score': peptide_mean_score,
            'vote_percentage': peptide_mean_vote
        }
    )

first_rank = pd.DataFrame(
    combined_scores
)
```
